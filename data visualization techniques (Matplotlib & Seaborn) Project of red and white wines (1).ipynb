**What patterns or trends can you observe from the distributions? Are there any features with skewed

```python
features = [
    'fixed acidity', 'volatile acidity', 'residual sugar', 'chlorides',
    'free sulfur dioxide', 'total sulfur dioxide', 'density',
    'pH', 'sulphates', 'alcohol'
]
# Calculate skewness and kurtosis
skew_kurt = pd.DataFrame({
    'Feature': features,
    'Skewness': [df_wine[feature].skew() for feature in features],
    'Kurtosis': [df_wine[feature].kurtosis() for feature in features]
})
print(skew_kurt)
# Bar plot for skewness
plt.figure(figsize=(10, 5))
plt.bar(skew_kurt['Feature'], skew_kurt['Skewness'], color='darkblue', edgecolor='black')
plt.xticks(rotation=45)
plt.title('Skewness of Wine Features')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.ylabel('Skewness')
plt.tight_layout()
plt.show()
```

**Task 5: Visualizing Distribution of Key Variables with Boxplots and Histograms
Goal: Explore relationships between wine quality and several key chemical features.
Task:
Visualize the distribution of key variables (alcohol, sulphates, citric acid, and fixed_acidity) with boxplots
and histograms.
What do you observe about how wine quality is related to these features? Are there any patterns that
stand out? note: Create a subplot showing boxplots and histograms for each of the selected features
(alcohol, sulphates, citric_acid, and fixed_acidity) with a title describing the feature being visualized.**

```python
plt.figure(figsize=(18, 10))
selected_features = ['alcohol', 'sulphates', 'citric acid', 'fixed acidity']
# Loop through each feature and create a boxplot and histogram
for i, feature in enumerate(selected_features, 1):
    # Boxplot (subplot 1)
    plt.subplot(2, 4, i)  # 2 rows, 4 columns, feature position
    sns.boxplot(x=df_wine[feature])
    plt.title(f'Boxplot of {feature}')
    
    # Histogram (subplot 2)
    plt.subplot(2, 4, i+4)  # 2 rows, 4 columns, feature position (second row)
    sns.histplot(df_wine[feature], kde=True, bins=15, color='green')
    plt.title(f'Histogram of {feature}')

plt.tight_layout()
plt.show()
```

**Task 8: Boxplot for Quality vs Various Features
Goal: Explore the impact of several features on wine quality.
Task:

density ), create a boxplot showing how the feature values vary across different wine quality ratings.

```python
features = ['fixed acidity', 'residual sugar', 'sulphates', 'citric acid', 'density']

# Set up the plot
plt.figure(figsize=(16, 18))

for idx, feature in enumerate(features, 1):
    plt.subplot(3, 2, idx)
    sns.boxplot(x='quality', y=feature, hue='quality', data=df_wine, palette='coolwarm')
    plt.title(f'{feature.capitalize()} vs Wine Quality')
    plt.xlabel('Wine Quality')
    plt.ylabel(feature.capitalize())

plt.tight_layout()
plt.show()
```
