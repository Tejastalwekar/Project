**Q5 Basic Data Analysis:
Calculate average salary by department.
Calculate the average performance score per department.
Identify employees with high performance but low salaries and vice versa.**

```python
df.groupby(["Department"]).agg({"Performance_Score":np.median})
```

```python
df.groupby(["Department"]).agg({"Salary":np.median})
```

```python
# High performance but low salary
high_performance_low_salary = df[(df['Performance_Score'] > 5) & (df['Salary'] < 70000)]

# Low performance but high salary
low_performance_high_salary = df[(df['Performance_Score'] < 6) & (df['Salary'] > 50000)]

# Display the results
print("\nEmployees with High Performance but Low Salary:")
print(high_performance_low_salary)

print("\nEmployees with Low Performance but High Salary:")
print(low_performance_high_salary)
```
